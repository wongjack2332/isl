```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('../all_csv/college.csv', index_col=0)
```

```python
df.head()
```

```python
df1 = df[['Top10perc', 'Apps', 'Enroll']]
df1.head()
```

```python
pd.plotting.scatter_matrix(df1)
```

```python
fig, axs = plt.subplots(1,2)
df.boxplot('Outstate', ax=axs[0])
df1 = df.loc[df['Private'] == 'Yes']
df1.boxplot('Outstate', ax=axs[1])
```

```python
fig, axs = plt.subplots(1,2)
df['Elite'] = pd.cut(df['Top10perc'],
[0, 50, 100],
labels=['No', 'Yes'])
df.boxplot('Outstate', ax=axs[0])
df1 = df.loc[df['Elite'] == 'Yes']
df1.boxplot('Outstate', ax=axs[1])
```
