```python
from matplotlib import pyplot as plt

fig, axs = plt.subplots(2,2)

df.plot.scatter(x='mpg', y='weight', ax=axs[0,0]) # higher the weight, lower the mpg
df.plot.scatter(x='mpg', y='year', ax=axs[0,1]) # older the car, higher the mpg
df.plot.scatter(x='mpg', y='displacement', ax=axs[1,0]) # loosely, higher the acceleration, higher the mpg
# the higher the displacement, the lower the mpg
```
